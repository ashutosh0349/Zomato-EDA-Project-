```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing dataset

```python
df=pd.read_csv('Zomato data .csv')
df
```

## Covert the data Type of rate Column

```python
def rate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)
```

```python
df['rate'] = df['rate'].apply(rate)
df
```

## Type of resturant that receive more offline orders

```python
# Create pivot table
pivot_table = df.pivot_table(
    index='listed_in(type)', 
    columns='online_order', 
    aggfunc='size',   # counts occurrences
    fill_value=0
)

# Plot heatmap
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap')
plt.xlabel('Mode of Order')
plt.ylabel('Listed In (Types)')
plt.show()
```

Conclusion - Dinning Type of resturant receive more offline order
